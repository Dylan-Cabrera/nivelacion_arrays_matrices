```python
personas = []

print("Ingresar el nombre, edad y nota de personas par ordenarlas por nota")

while True:
    repetir = input("¿Crear nuevo registro? Y/n").lower()
    if repetir == "n":
        break

    nombre = input("Nombre:")
    edad = int(input("Edad:"))
    nota = int(input("Nota:"))

    persona = [nombre, edad, nota]

    personas.append(persona)

print("Personas ingresadas:")
for p in personas:
    print(f" \n Nombre: {p[0]} \n Edad: {p[1]} \n Nota: {p[2]}")

personas.sort(reverse=True, key= lambda p: p[2])

print("Lista ordenada")
for p in personas:
    print(f"\n Nombre: {p[0]} \n Edad: {p[1]} \n Nota: {p[2]}")

```
