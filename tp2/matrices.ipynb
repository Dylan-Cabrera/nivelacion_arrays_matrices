```python
personas = []
total_notas = 0
i = 0

print("Ingresar el nombre, edad y nota de personas para ordenarlas por nota")

while True:
    repetir = input("¿Crear nuevo registro? Y/n").lower()
    if repetir == "n":
        break

    i += 1
    nombre = input("Nombre:")
    edad = int(input("Edad:"))
    nota = int(input("Nota:"))
    total_notas += nota

    persona = [nombre, edad, nota]

    personas.append(persona)

print("Personas ingresadas:")
for p in personas:
    print(f" \n Nombre: {p[0]} \n Edad: {p[1]} \n Nota: {p[2]}")

personas.sort(reverse=True, key= lambda p: p[2])

print("Lista ordenada")
for p in personas:
    print(f"\n Nombre: {p[0]} \n Edad: {p[1]} \n Nota: {p[2]}")

print(f"\n Promedio de notas general: {total_notas/i}")

```
